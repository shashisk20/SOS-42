```python
!pip install gTTS

from gtts import gTTS
import IPython.display as ipd

# Input text
text = "Type your text here"

# Create a gTTS object
tts = gTTS(text, lang='en')

# Save the audio file
audio_path = '/content/output.mp3'
tts.save(audio_path)

# Display the audio
ipd.Audio(audio_path)
```
